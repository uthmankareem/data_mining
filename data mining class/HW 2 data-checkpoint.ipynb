```python
#Question 1 for normalization function
data1 = data
data1 = data1.drop("class", axis = 1)
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        mean_value = df[feature_name].mean()
        std_value = df[feature_name].std()
        result[feature_name] = (df[feature_name] - mean_value) / std_value
    return result
#test
norm1=normalize(data1)

norm1.head()
```

```python
#setosa correlation after normalization
setosanorm=normalize(c.drop("class", axis=1))
print(setosanorm.corr())

#versicolor correlation after normalization
versicolornorm = normalize(d.drop("class", axis=1))
print(versicolornorm.corr())
#verginica correlation after normalization
virginicanorm = normalize(e.drop("class", axis=1))
print(virginicanorm.corr())
```

```python
#Question 1 minkowski function
def minkoski(A, B, r):
    A = np.array([A])
    mink = round(np.sum(abs(A - B)**r)**(1/r), 2)
    return mink
```

```python
#Question 2 T index funtion
def T_sta(A,B):
    t = (np.mean(A) - np.mean(B))/np.sqrt((np.var(A) + np.var(B))/len(A))
    return abs(t)
```

```python
#Question 3 mahalanobis distance function
def mal(A, B, M):
    mah = np.sqrt(abs((A - B)@(np.linalg.inv(M))@(A - B).transpose()))
    return mah
```

```python
#Question 1 - minkowski distance calculation for S
S = [4.50, 3.50, 1.46, 0.24]
m=data1.to_numpy()
net =[]
for i in range(len(m)):
    b=m[i]
    net.append(minkoski(S, b, 1))
print(net)
#index with nearest distance is index with minmum value
print(net.index(min(net)))
# index 6 has the minimum distance
 

#plotting the distance obtained
plt.figure()
plt.plot(net)
plt.show()
```

```python
#Question 2 mahalanobis distance caclulation for S
M = data1.cov()
malh = []
for i in range(len(m)):
    b=m[i]
    malh.append(mal(S, b, M))
print(malh)
#index with nearest distance is index with minmum value
print(malh.index(min(malh)))
# index 11 has the closest distance to S
 

#plotting the distance obtained
plt.figure()
plt.plot(malh)
plt.show()
len(malh)
```
